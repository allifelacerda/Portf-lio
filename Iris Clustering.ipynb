Estudando Algoritmo KMeans

O dataset Iris é comumente usado para demonstração de algoritmos de machine learning. Isso é um breve estudo utilizando-o.

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.cluster import KMeans
%matplotlib inline
```

### Carregando a base de dados iris

```python
df = pd.read_csv("iris.csv")
```

```python
df.head()
```

### Visualizando o relacionamento de todas as variáveis

```python
sns.pairplot(df)
```

### Isolando as colunas das classes, afinal, queremos clusterizar as classes.

```python
X = np.array(df.drop('Species', axis = 1))
```

```python
X[:5]
```

### Dividindo os dados em 3 classes utilizando KMeans

```python
kmeans = KMeans(n_clusters=3, random_state=0)
```

```python
kmeans.fit(X)
```

```python
kmeans.labels_
```

### Comparando as classes criadas pelo algoritmo KMeans com as classes originais dos dados

```python
df['KMeans - Classes'] = kmeans.labels_
```

```python
df.head()
```

### Visualizando o relacionamento de todas as variáveis novamente, comparando os dois conjuntos de dados.

```python
sns.pairplot(df.drop('KMeans - Classes', axis=1), hue = 'Species')
```

```python
sns.pairplot(df.drop('Species', axis=1), hue = 'KMeans - Classes')
```

```python
df['Species'].value_counts()
```

```python
df['KMeans - Classes'].value_counts()
```

### Podemos notar visualizando os gráficos uma grande quantidade de acertos de classificação.
